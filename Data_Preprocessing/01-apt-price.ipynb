### 1.4.5 데이터 타입 변경

```python
# 에러 발생 , 데이터 내의 공백문자 parsing 문제
df_last['분양가격'] = pd.to_numeric(df_last['분양가격(㎡)'], errors='coerce')
df_last['분양가격'].mean()
df_last['분양가격'].max()
```

### 1.4.6 평당분양가격 구하기

```python
df_last['평당분양가격'] = df_last['분양가격'] * 3.3
df_last.head(1)
```

### 1.4.8 규모구분을 전용면적 컬럼으로 변경

```python
df_last['규모구분'].unique()
```

```python
## 중요한 technique , 평소에 많이 헷갈려했던 것 !
df_last['전용면적'] = df_last['규모구분'].str.replace('전용면적', '')
df_last['전용면적'] = df_last['전용면적'].str.replace('초과', '~')
df_last['전용면적'] = df_last['전용면적'].str.replace('이하', '')
df_last['전용면적'] = df_last['전용면적'].str.replace(' ', '')
df_last['전용면적'] = df_last['전용면적'].str.strip()
df_last['전용면적']
```

### 1.8.1 melt 로 data 합치기 ( 진짜 좋은 테크닉 !!! 꼭 기억하쇼 !!!)

### 1.8.2 연도와 월을 분리하기

```python
date = '2013년12월'
date
```

```python
# '년'을 기준으로 텍스트를 분리해 봅니다.
date.split('년')
```

```python
date.split('년')[0]
```

```python
date.split('년')[-1].replace('월', '')
```

```python
def parse_year(date):
    year = date.split('년')[0]
    year = int(year)
    return year

y = parse_year(date)
print(type(y))
y
```

```python
parse_year(date)
```

```python
def parse_month(date):
    month = date.split('년')[-1].replace('월', '')
    month = int(month)
    return month

month = parse_month(date)
print(type(month))
month
```

```python
## 핵 중요한 테크닉!!!!
df_first_melt['연도'] = df_first_melt['기간'].apply(parse_year)
df_first_melt['월'] = df_first_melt['기간'].apply(parse_month)
```

```python
df_first_melt.head()
```

```python
df_last.columns.to_list()
```
