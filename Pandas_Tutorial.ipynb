## Data Cleaning/Manipulation

### Apply customized Functions.

```python
df2['type'].value_counts()
```

What if we want to perform different operations on different tyhpes of transactions;

    - when it's a deposit, calculate amount/balance.
    
    - when it's a withdrawal, calculate amount/(balance + amount).

```python
def pct_of_balance_trans(row):
    if row['type'] == 'DEPOSIT':
        return row['amount']/row['balance']
    elif row['type'] == 'WITHDRWAL' :
        return row['amount']/(row['balance'] + row['amount'])
    else : 
        -999
```

```python
df2['pct_of_balance_trans'] = df2.apply(pct_of_balance_trans, axis=1)

df2
```

## Merge and Concatenate DataFrames

Simliar to SQL tables, we can also merge or concatenate DataFrames.

```python
left = pd.DataFrame({'col1': [1,2,3,4,5], 'col2': ['a', 'b', 'c', 'd', 'e']})
right = pd.DataFrame({'col1': [3,4,5,6,7], 'col3': ['f', 'g', 'h', 'i', 'j']})

print(left)
print(right)
```

### Merge

The 'merge' join is done on columns or indexes.

For example, we can merge the left and right DataFrames on their common column col1. 

The how parameter specifies the type of merge to be performed.

when how = 'left', the new DataFrame keeps all the rows from the left DataFrame and the corresponding values from the DataFrame.

```python
pd.merge(left, right, on=['col1'], how='inner')

pd.merge(left, right, on=['col1'], how='left')

pd.merge(left, right, on=['col1'], how='right')

pd.merge(left, right, on=['col1'], how='outer')
```

### Concatenate

When we have two DataFrames with the same row or column information, we can glue them together using 'concat'.

```python
d = {'account_id': [1, 2, 3, 11382], 'gender': ['female', 'male', 'female', 'male'], 'age': [55, 25, 29, 39]}
df4 = pd.DataFrame(d)
df4

d = {'account_id': [4, 5, 6, 7], 'gender': ['female', 'male', 'female', 'male'], 'age': [12, 13, 14, 15]}
df5 = pd.DataFrame(d)
df5
```

```python
df6 = pd.concat([df4, df5])
df6
```
